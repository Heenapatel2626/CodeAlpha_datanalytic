```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("master.csv")
print("---Data INFO---")
print(df.info())
print("\n---FIRST 5 ROWS---")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('master.csv')

plt.figure(figsize=(6, 6))
gender_counts = df.groupby('sex')['suicides_no'].sum()

plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#4C72B0', '#DD8452'],
    explode=(0.05, 0),
)

plt.title('Suicides Share by Gender', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('gender_pie.png')
plt.show()

plt.figure(figsize=(7, 7))
gen_counts = df.groupby('generation')['suicides_no'].sum()

plt.pie(
    gen_counts,
    labels=gen_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Set3.colors,
)

plt.title('Suicides Share by Generation', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('generation_pie.png')
plt.show()
```
