```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('master.csv')
sns.set_theme(style="whitegrid")

plt.figure(figsize=(10, 5))
yearly_suicides = df.groupby('year')['suicides_no'].sum().reset_index()
yearly_suicides = yearly_suicides[yearly_suicides['year'] < 2016]

sns.lineplot(data=yearly_suicides, x='year', y='suicides_no', marker='o', color='b', linewidth=2.5)
plt.title('Global Suicide Numbers Over Years (1985 - 2015)', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Suicides', fontsize=12)
plt.tight_layout()
plt.savefig('global_trend.png')
plt.show()

plt.figure(figsize=(10, 5))
age_gender = df.groupby(['age', 'sex'])['suicides_no'].sum().reset_index()
age_order = ['5-14 years', '15-24 years', '25-34 years', '35-54 years', '55-74 years', '75+ years']
age_gender['age'] = pd.Categorical(age_gender['age'], categories=age_order, ordered=True)
age_gender = age_gender.dropna().sort_values('age')

sns.barplot(data=age_gender, x='age', y='suicides_no', hue='sex', palette='muted')
plt.title('Total Suicides by Age Group and Gender', fontsize=14, fontweight='bold')
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Total Suicides', fontsize=12)
plt.tight_layout()
plt.savefig('age_gender_distribution.png')
plt.show()
```
